```python
# TASK 3: Calculate distances between a launch site and nearby points (no geopy needed)

import math
import folium

def haversine_km(p1, p2):
    """Great-circle distance between two [lat, lon] points in kilometers."""
    lat1, lon1 = map(math.radians, p1)
    lat2, lon2 = map(math.radians, p2)
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    return 6371.0088 * (2 * math.asin(math.sqrt(a)))  # Earth's mean radius (km)

# Example: pick one launch site (CCAFS SLC-40)
launch_site = [28.562302, -80.577356]

# Example proximity points (adjust to what your lab specifies)
coastline    = [28.563,  -80.567]      # approximate coast
orlando_city = [28.5383, -81.3792]     # Orlando, FL
highway      = [28.5721, -80.5853]     # nearby highway coordinate

# Distances
dist_coast   = haversine_km(launch_site, coastline)
dist_city    = haversine_km(launch_site, orlando_city)
dist_highway = haversine_km(launch_site, highway)

print(f"Distance to Coastline: {dist_coast:.2f} km")
print(f"Distance to Orlando:   {dist_city:.2f} km")
print(f"Distance to Highway:   {dist_highway:.2f} km")

# Visualize on a Folium map
site_map = folium.Map(location=launch_site, zoom_start=10)

# Launch site
folium.Marker(launch_site, popup='CCAFS SLC-40', icon=folium.Icon(color='blue')).add_to(site_map)

# Coastline
folium.Marker(coastline, popup=f'Coastline ({dist_coast:.2f} km)', icon=folium.Icon(color='green')).add_to(site_map)
folium.PolyLine([launch_site, coastline], color='green').add_to(site_map)

# Orlando
folium.Marker(orlando_city, popup=f'Orlando ({dist_city:.2f} km)', icon=folium.Icon(color='red')).add_to(site_map)
folium.PolyLine([launch_site, orlando_city], color='red').add_to(site_map)

# Highway
folium.Marker(highway, popup=f'Highway ({dist_highway:.2f} km)', icon=folium.Icon(color='purple')).add_to(site_map)
folium.PolyLine([launch_site, highway], color='purple').add_to(site_map)

site_map  # If blank, File  Trust Notebook, then re-run
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition, MeasureControl

# (Re)use your existing site_map, or create one if needed:
# site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=10)

# 1) Live lat/lon readout on hover
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' | ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Coordinates:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# 2) Interactive distance ruler (click to measure; Esc/double-click to finish)
site_map.add_child(MeasureControl(
    position='topleft',
    primary_length_unit='kilometers',   # 'meters','kilometers','miles','feet'
    secondary_length_unit='miles',
    primary_area_unit='sqmeters'        # not really used here, but ok to keep
))

# Render + optional save
site_map
site_map.save("site_map_with_mouse_and_ruler.html")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Example: find coordinate of the closest coastline (from MousePosition hover)
coastline_lat, coastline_lon = 28.56367, -80.57163   # <-- replace with your picked coords

# Define Haversine distance function (if not defined already)
import math
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0088  # Earth radius in km
    1, 2 = math.radians(lat1), math.radians(lat2)
    d = math.radians(lat2 - lat1)
    d = math.radians(lon2 - lon1)
    a = math.sin(d/2)**2 + math.cos(1)*math.cos(2)*math.sin(d/2)**2
    return R * (2 * math.asin(math.sqrt(a)))

# Calculate distance between launch site and this coastline point
distance_coastline = calculate_distance(
    launch_site[0], launch_site[1], coastline_lat, coastline_lon
)
distance_coastline
```
